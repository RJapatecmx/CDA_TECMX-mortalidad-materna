```python
import pandas as pd

df = pd.read_csv("maternal_mortality.csv")

df.head()
```

```python
mmr_cols = [col for col in df.columns if "Maternal Mortality Ratio" in col]
df_long = df.melt(
    id_vars=['ISO3', 'Country', 'Continent', 'Hemisphere',
             'Human Development Groups', 'UNDP Developeing Regions',
             'HDI Rank (2021)'],
    value_vars=mmr_cols,
    var_name='Year',
    value_name='MMR'
)

df_long.head()
```

```python
df_long['Year'] = df_long['Year'].str.extract(r'(\d{4})')
df_long['Year'] = df_long['Year'].astype(int)

df_long.head()
```

```python
df_long.shape
df_long.info()
```

```python
df_long['MMR'].describe()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.hist(df_long['MMR'], bins=30)
plt.title("DistribuciÃ³n de la Mortalidad Materna")
plt.xlabel("MMR")
plt.ylabel("Frecuencia")
plt.show()
```

La distribuciÃ³n de la Mortalidad Materna presenta un fuerte sesgo positivo (right-skewed), con la mayorÃ­a de los paÃ­ses concentrados en niveles bajos, pero con una cola larga que refleja valores extremadamente altos en algunos paÃ­ses. Esto sugiere alta desigualdad global en resultados de salud materna.



```python
df_2021 = df_long[df_long['Year'] == 2021]

plt.figure()
plt.hist(df_2021['MMR'], bins=30)
plt.title("DistribuciÃ³n de MMR en 2021")
plt.xlabel("MMR")
plt.ylabel("Frecuencia")
plt.show()
```

```python
continent_stats = df_long.groupby('Continent')['MMR'].agg(['mean','median','std','min','max'])
continent_stats
```

La mortalidad materna presenta fuertes disparidades regionales. Ãfrica presenta niveles considerablemente mÃ¡s altos que Europa, con promedios casi 45 veces mayores. Asia muestra una heterogeneidad interna considerable, como lo indica la diferencia entre los valores promedio y mediano.

```python
plt.figure()
df_long.boxplot(column='MMR', by='Continent', rot=45)
plt.title("DistribuciÃ³n de MMR por Continente")
plt.suptitle("")
plt.ylabel("MMR")
plt.show()
```

```python
hdi_stats = df_long.groupby('Human Development Groups')['MMR'].agg(['mean','median','std'])
hdi_stats
```

La mortalidad materna parece estar estrechamente relacionada con los niveles de desarrollo humano. Los paÃ­ses clasificados como IDH bajo presentan tasas de mortalidad casi 40 veces superiores a las de los paÃ­ses con IDH muy alto, lo que sugiere que los factores estructurales del desarrollo podrÃ­an ser mÃ¡s determinantes que la ubicaciÃ³n geogrÃ¡fica.


```python
df_long['Period'] = pd.cut(
    df_long['Year'],
    bins=[1989, 1999, 2009, 2021],
    labels=['1990s', '2000s', '2010s']
)

df_long.groupby('Period')['MMR'].mean()
```

```python
df_long.groupby(['Continent','Period'])['MMR'].mean()
```

```python
top10 = df_long.groupby('Country')['MMR'].mean().sort_values(ascending=False).head(10)
top10
```

```python
continent_period = df_long.groupby(['Continent','Period'])['MMR'].mean().unstack()
continent_period['Change_%'] = ((continent_period['2010s'] - continent_period['1990s']) / continent_period['1990s']) * 100
continent_period
```
